```python
import pandas as pd
import numpy as np
import os
```

# Adquisición de data

```python
jsonList = sorted(os.listdir('../json/'))
len(jsonList)
```

```python
# 
```

```python
data_df = pd.read_json('../json/datos_2024-01-01.json')
data_df.head(1)
```

```python
for json_doc in jsonList[1:]:
    data_df = pd.concat([data_df,pd.read_json(f'../json/{json_doc}')],axis=0)
```

```python
data_df.info()
```

```python
pd.value_counts(['<p>' in x for x in data_df['summary'] if x is not None])
```

```python
data_df['summary'] = [x.replace('</p>','').replace('<p>','') if x is not None else x for x in data_df['summary']]
```

```python
pd.unique([len(x.keys()) for x in data_df['rating']])
```

```python
data_df['rating'] = [x['average'] for x in data_df['rating']]
```

```python
data_df
```

```python
[x for x in data_df['_links']]
```

```python
pd.unique([len(x) for x in data_df['_embedded']]), pd.unique([len(x['show']) for x in data_df['_embedded']])
```

```python
data_df['genres'] = [','.join(x['show']['genres']) for x in data_df['_embedded']]
```

```python
len([x['show']['officialSite'] for x in data_df['_embedded']])
```

```python
data_df['officialSite'] = [x['show']['officialSite'] for x in data_df['_embedded']]
data_df['nameShow'] = [x['show']['name'] for x in data_df['_embedded']]
data_df['averageRuntime'] = [x['show']['averageRuntime'] for x in data_df['_embedded']]
```

```python
data_df
```

```python
data_df.drop(columns=['_links','_embedded','image'],inplace=True)
```

```python
data_df.to_parquet('../data/data.parquet',compression='snappy',index=False)
```
